```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('coords.csv')
coords = data.values.tolist()
```

```python
# Axis parameters
xmin = 0
xmax = 21
ymin = 0
ymax = 10
```

```python
# Create gridpoints
xgrid = []
ygrid = []
for x in range(xmin, xmax, 1):
    for y in range(ymin, ymax, 1):
        xgrid.append(x)
        ygrid.append(y)
```

```python
for i in range(len(coords)):
    # Turn x coords into floats
    x1 = float(coords[i][0][1:]) 
    x2 = float(coords[i][1][1:])
    # Turn y coords into integers
    y1 = ord(coords[i][0][0]) - 65
    y2 = ord(coords[i][1][0]) - 65
    # Plot the two points with a point and join them
    plt.plot([x1,x2], [y1,y2], 'o-')
# Flip the y axis so it goes A-J
plt.gca().invert_yaxis()
# Set the labels / frequency of ticks
plt.yticks(ticks=range(ymin,ymax,1), labels=['J', 'I', 'H', 'G', 'F', 'E', 'D', 'C', 'B', 'A'])
plt.xticks(ticks=range(xmin,xmax,2))
# Remove the ticks from the axes and add the labels to the left and top
plt.tick_params(top=False, labeltop=True, bottom=False, labelbottom=False, labelleft=True, left=False)
# Remove the axis splines
plt.box()
plt.plot(xgrid,ygrid,'.')
plt.show()
```
